```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from osgeo import gdal
```

# SM time series

```python
folder=r'E:\rahul\processed_data\soil moisture'
```

```python
SM_PATH=[]
for root, dirs, files in os.walk(folder):
    for names in files:
        full_path=os.path.join(root,names)
        SM_PATH.append(full_path)
```

```python
profile1=[]
profile2=[]
profile3=[]
profile4=[]
profile5=[]
profile6=[]
```

```python
for i in range(30):
    data=gdal.Open(SM_PATH[i])
    data=data.ReadAsArray()
    profile1.append((data[40,16]).round(4))
    profile2.append((data[51,37]).round(4))
    profile3.append((data[50,74]).round(4))
    profile4.append((data[17,40]).round(4))
    profile5.append((data[72,39]).round(4))
    profile6.append((data[84,25]).round(4))
```

```python
df=(pd.DataFrame([profile1,profile2,profile3,profile4,profile5,profile6])).T
```

```python
df
```

```python
df.to_csv('E:/rahul/processed_data/SM.csv')
```

# BT_10V

```python
folder=r'E:\rahul\processed_data\BT_10V'
```

```python
BT_PATH=[]
for root, dirs, files in os.walk(folder):
    for names in files:
        full_path=os.path.join(root,names)
        BT_PATH.append(full_path)
```

```python
profile1=[]
profile2=[]
profile3=[]
profile4=[]
profile5=[]
profile6=[]
for i in range(30):
    BT_10V=gdal.Open(BT_PATH[i])
    BT_10V=(BT_10V.ReadAsArray().astype('float64'))/10
    BT_10V[BT_10V<=100.0]=np.nan
    BT_10V[BT_10V>=500.0]=np.nan
    profile1.append((BT_10V[40,16]).round(4))
    profile2.append((BT_10V[51,37]).round(4))
    profile3.append((BT_10V[50,74]).round(4))
    profile4.append((BT_10V[17,40]).round(4))
    profile5.append((BT_10V[72,39]).round(4))
    profile6.append((BT_10V[84,25]).round(4))
```

```python
df=(pd.DataFrame([profile1,profile2,profile3,profile4,profile5,profile6])).T
```

```python
df.to_csv('E:/rahul/processed_data/BT_10V.csv')
```
