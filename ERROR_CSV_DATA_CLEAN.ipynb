```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
path=r'E:/23sept/Error_csv/'
path_file=[]
for root, dirs, files in os.walk(path):
    for name in files:
        print(path+name)
        path_file.append(path+name)
```

```python
df=pd.DataFrame(columns=['DOY', 'Wavelenght', 'MODIS_LST', 'LST_corrected',
       'LST_raw', 'Raw_LST_error', 'Corrected_LST_error'])
for i in range(len(path_file)):
    data=pd.read_csv(path_file[i])
    data.dropna(inplace=True)
    data=data[(data['Wavelenght']!=36) & (data['Wavelenght']!=89)]
    data['Raw_LST_error']=data['LST_raw']-data['MODIS_LST']
    data['Corrected_LST_error']=data['LST_corrected']-data['MODIS_LST']
    data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
    df=df.append(data,ignore_index=True)
    
```

```python
df.to_csv(r'E:23sept/Error_plot/Cleanned_data_all.csv')
```

```python
Error_E1=df[['Raw_LST_error', 'Corrected_LST_error']]
plt.figure(figsize=(10,7))
plt.plot(Error_E1)
plt.ylabel("LST Error (K)")
plt.title('LST_ERROR_PLOT',fontweight ="bold")
plt.legend(['Raw_LST_error', 'Corrected_LST_error'],loc='upper left')
plt.savefig(r'E:23sept/Irradiance_LST_error/Error.png',dpi=300)
plt.show()
```
