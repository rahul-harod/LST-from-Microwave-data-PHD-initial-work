```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
data=pd.ExcelFile('E:\MYD11A1_Data.xlsx')
```

```python
names=data.sheet_names
names
```

```python
sheet_number=0
```

```python
file=pd.read_excel('E:\MYD11A1_Data.xlsx',sheet_number,na_values=['#VALUE!', '#DIV/0!'])
file
```

```python
file.set_index('Date', inplace=True)
```

```python
file_Emiss_1KM=file[["Ground_LST_1KM_E1","MYD11A1_1KM_LST","MYD11A1_6KM_LST","MYD11C1_6KM_LST","A1_1KM_E1","A1_6KM_E1","C1_6KM_E1"]]
file_Emiss_1KM
```

```python
file_Emiss_1KM.describe()
```

```python
file_Emiss_1KM.dropna(inplace=True)
```

```python
file_Emiss_1KM=file_Emiss_1KM[(file_Emiss_1KM['A1_1KM_E1']>-10) & (file_Emiss_1KM['A1_1KM_E1']<10) & (file_Emiss_1KM['A1_6KM_E1']>-10) & (file_Emiss_1KM['A1_6KM_E1']<10) & (file_Emiss_1KM['C1_6KM_E1']>-10) &(file_Emiss_1KM['C1_6KM_E1']<10)]
```

```python
file_Emiss_1KM.describe()
```

```python
Error_E1=file_Emiss_1KM[["A1_1KM_E1","A1_6KM_E1","C1_6KM_E1"]]
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(data=Error_E1)
plt.ylabel("LST Error (K)")
plt.title('Site: '+ str(names[sheet_number]),fontweight ="bold")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.violinplot(data=file,inner='quartile',linewidth=2)
sns.violinplot(data=file,inner='box',linewidth=3)
plt.ylabel("LST Error (K)")
plt.title('Site: '+ str(names[sheet_number]),fontweight ="bold")
plt.show()
```
