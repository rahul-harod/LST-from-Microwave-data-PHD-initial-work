```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from osgeo import gdal
```

# 1. file path

```python
# AMSER-E BT 
folder_BT=r'E:\rahul\processed_data\MAY_2007\ASTER_BT_clipped'
f_BT06_9V=[]
f_BT10_7V=[]
f_BT18_7V=[]
f_BT36_5V=[]
f_BT89_0V=[]
f_BT06_9H=[]
f_BT10_7H=[]
f_BT18_7H=[]
f_BT36_5H=[]
f_BT89_0H=[]
f_soilM=[]
f_vegW=[]

# MODIS LST
folder_LST=r'E:\rahul\processed_data\MAY_2007\MODIS_LST_clipped'
f_LST=[]


for root, dirs, files in os.walk(folder_BT):
    for names in files:
        if 'TB06_9V' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_BT06_9V.append(full_path)
            
        if 'TB10_7V' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_BT10_7V.append(full_path)
            
        if 'TB18_7V' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_BT18_7V.append(full_path)
            
        if 'TB36_5V_Res_1' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_BT36_5V.append(full_path)
            
        if 'TB89_0V' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_BT89_0V.append(full_path)
            
        if 'TB06_9H' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_BT06_9H.append(full_path)
            
        if 'TB10_7H' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_BT10_7H.append(full_path)
            
        if 'TB18_7H' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_BT18_7H.append(full_path)
            
        if 'TB36_5H_Res_1' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_BT36_5H.append(full_path)
            
        if 'TB89_0H' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_BT89_0H.append(full_path)
            
        if 'A_Soil_Moisture' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_soilM.append(full_path)
            
        if 'A_Veg_Water_Content' in names:
            full_path=os.path.join(root,names)
            print(full_path)
            f_vegW.append(full_path)
            

for root, dirs, files in os.walk(folder_LST):
    for names in files:
        full_path=os.path.join(root,names)
        print(full_path)
        f_LST.append(full_path)
        
```

```python
len(f_BT06_9H)
```

```python
# Crop the dataset[A:B,C:D]
# -----------------------
# # # North INDIA
# # rows
# A=20
# B=47
# # columns
# C=8
# D=52
# -----------------------
# # Central INDIA
# # rows
# A=48
# B=65
# # # columns
# C=26
# D=52
# -----------------------
# # South INDIA
# rows
A=77
B=106
# columns
C=22
D=49
```

```python
BT_6V=[]
BT_6H=[]
BT_10V=[]
BT_10H=[]
BT_18V=[]
BT_18H=[]
BT_36V=[]
BT_36H=[]
BT_89V=[]
BT_89H=[]
SM_10V=[]
VegW=[]
lst=[]
for i in range(len(f_LST)):
    BT=gdal.Open(f_BT06_9H[i])
    BT=(BT.ReadAsArray().astype('float64'))/10
    BT[BT<=100.0]=np.nan
    BT[BT>=400.0]=np.nan
    BT_6H.append((BT[A:B,C:D]).reshape(-1))
    
    BT=gdal.Open(f_BT06_9V[i])
    BT=(BT.ReadAsArray().astype('float64'))/10
    BT[BT<=100.0]=np.nan
    BT[BT>=400.0]=np.nan
    BT_6V.append((BT[A:B,C:D]).reshape(-1))
    
    BT=gdal.Open(f_BT10_7H[i])
    BT=(BT.ReadAsArray().astype('float64'))/10
    BT[BT<=100.0]=np.nan
    BT[BT>=400.0]=np.nan
    BT_10H.append((BT[A:B,C:D]).reshape(-1))
    
    BT=gdal.Open(f_BT10_7V[i])
    BT=(BT.ReadAsArray().astype('float64'))/10
    BT[BT<=100.0]=np.nan
    BT[BT>=400.0]=np.nan
    BT_10V.append((BT[A:B,C:D]).reshape(-1))
    
    BT=gdal.Open(f_BT18_7H[i])
    BT=(BT.ReadAsArray().astype('float64'))/10
    BT[BT<=100.0]=np.nan
    BT[BT>=400.0]=np.nan
    BT_18H.append((BT[A:B,C:D]).reshape(-1))
    
    BT=gdal.Open(f_BT18_7V[i])
    BT=(BT.ReadAsArray().astype('float64'))/10
    BT[BT<=100.0]=np.nan
    BT[BT>=400.0]=np.nan
    BT_18V.append((BT[A:B,C:D]).reshape(-1))
    
    BT=gdal.Open(f_BT36_5H[i])
    BT=(BT.ReadAsArray().astype('float64'))/10
    BT[BT<=100.0]=np.nan
    BT[BT>=400.0]=np.nan
    BT_36H.append((BT[A:B,C:D]).reshape(-1))
    
    BT=gdal.Open(f_BT36_5V[i])
    BT=(BT.ReadAsArray().astype('float64'))/10
    BT[BT<=100.0]=np.nan
    BT[BT>=400.0]=np.nan
    BT_36V.append((BT[A:B,C:D]).reshape(-1))
    
    BT=gdal.Open(f_BT89_0H[i])
    BT=(BT.ReadAsArray().astype('float64'))/10
    BT[BT<=100.0]=np.nan
    BT[BT>=400.0]=np.nan
    BT_89H.append((BT[A:B,C:D]).reshape(-1))
    
    BT=gdal.Open(f_BT89_0V[i])
    BT=(BT.ReadAsArray().astype('float64'))/10
    BT[BT<=100.0]=np.nan
    BT[BT>=400.0]=np.nan
    BT_89V.append((BT[A:B,C:D]).reshape(-1))
    
#     LST
    LST=gdal.Open(f_LST[i])
    LST=(LST.ReadAsArray().astype('float64'))
    lst.append((LST[A:B,C:D]).reshape(-1))
    
#     Soil moisture
    BT=gdal.Open(f_soilM[i])
    BT=(BT.ReadAsArray().astype('float64'))
    BT[BT<=0.0]=np.nan
    BT[BT>=500.0]=np.nan
    BT=BT*0.001
    SM_10V.append((BT[A:B,C:D]).reshape(-1))
    
#     vegetation water
    BT=gdal.Open(f_vegW[i])
    BT=(BT.ReadAsArray().astype('float64'))
    BT[BT<=0.0]=np.nan
    BT[BT>=500.0]=np.nan
    BT=BT*0.01
    VegW.append((BT[A:B,C:D]).reshape(-1))
```

```python
df=pd.DataFrame(columns=["BT_6V","BT_6H","BT_10V","BT_10H","BT_18V","BT_18H","BT_36V","BT_36H","BT_89V","BT_89H","MODIS_LST","SM","VegWater"])
for i in range(len(BT_6H)):
    df1=pd.DataFrame({
                        "BT_6V":BT_6V[i],"BT_6H":BT_6H[i],
                        "BT_10V":BT_10V[i],"BT_10H":BT_10H[i],
                        "BT_18V":BT_18V[i],"BT_18H":BT_18H[i],
                        "BT_36V":BT_36V[i],"BT_36H":BT_36H[i],
                        "BT_89V":BT_89V[i],"BT_89H":BT_89H[i],
                        "MODIS_LST":lst[i],"SM":SM_10V[i],"VegWater":VegW[i]
    })
    df=df.append(df1,ignore_index=True)
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df['PR6']=df['BT_6H']/df['BT_6V']
df['PR10']=df['BT_10H']/df['BT_10V']
df['PR18']=df['BT_18H']/df['BT_18V']
df['PR36']=df['BT_36H']/df['BT_36V']
df['PR89']=df['BT_89H']/df['BT_89V']
```

```python
LST_zero=df[df['MODIS_LST']==0.0].index
df.drop(LST_zero,inplace=True)
df.reset_index(drop=True, inplace=True)
```
