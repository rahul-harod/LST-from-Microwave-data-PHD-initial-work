```python
path='E:\MODINPUT'
for root, dirs, files in os.walk('E:\MODINPUT'):
    for name in files:
        print(name)
```

```python
output_name='North_east_APRIL_1_10'
initial_name_up=output_name+'_up'
initial_name_down=output_name+'_down'
filePath_up=[]
filePath_down=[]
for i in os.listdir(path):
    if os.path.isfile(os.path.join(path,i)) and 'scan.csv'  in i:
        if initial_name_up in i:
            filePath_up.append(path+'\\'+i)
        if initial_name_down in i:
            filePath_down.append(path+'\\'+i)
```

```python
filePath_up
```

```python
atm_contrib=pd.DataFrame(columns=['DOY','wavelen_um', 'Transmissivity',  'Upwelling_BT(K)','Downwelling_BT(K)','UP_radiance','UP_TOA_irradiance','DOWN_radiance','DOWN_TOA_irradiance'])
for i in range(len(filePath_up)):
#     UPwell
    df_up=pd.read_csv(filePath_up[0],header=[4,5])[:-1]
    df_up.columns = df_up.columns.map('_'.join)
    # df.dropna() 
    df_up=df_up.astype('float64')
    f_up=df_up[df_up['Waveln_  [um]'].isin([43300.0,28200.0,16000.0,12600.0,8200.0,3400.0])]
    f_up=f_up[['Waveln_  [um]','  combin_  path trans','  bbody_  temp[K]','  total_   rad','  ToA_  irrad']]
    f_up.rename(columns={"Waveln_  [um]": "wavelen_um", "  combin_  path trans": "Transmissivity",
                        '  bbody_  temp[K]':'Upwelling_BT(K)','  total_   rad':'UP_radiance',
                         '  ToA_  irrad':'UP_TOA_irradiance'},inplace=True)
    f_up['DOY']=(int(filePath_up[i][-11:-9].replace("_","")))
#     print(f_up)
    df_down=pd.read_csv(filePath_down[0],header=[4,5])[:-1]
    df_down.columns = df_down.columns.map('_'.join)
    # df.dropna() 
    df_down=df_down.astype('float64')
    f_down=df_down[df_down['Waveln_  [um]'].isin([43300.0,28200.0,16000.0,12600.0,8200.0,3400.0])]
    f_down=f_down[['Waveln_  [um]','  bbody_  temp[K]','  total_   rad','  ToA_  irrad']]
    f_down.rename(columns={'Waveln_  [um]':'wavelen_um','  bbody_  temp[K]':'Downwelling_BT(K)',
                           '  total_   rad':'DOWN_radiance','  ToA_  irrad':'DOWN_TOA_irradiance'},inplace=True)
#     print(f_down)
    result = pd.merge(f_up, f_down)
    atm_contrib=atm_contrib.append(result,ignore_index=True)

```
